```python
from scipy.io import netcdf
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
nc = netcdf.netcdf_file("data/era-int_pl_1979-2019-mmzm-u.nc")
```

```python
level = -15

u = nc.variables["u"][:, level, :, 0]
u.shape
```

```python
plt.figure(figsize = (18,8))

sns.heatmap(np.transpose(u), cmap="YlGnBu", xticklabels=48, yticklabels=50)

plt.ylabel("Latitude")
plt.xlabel("Time (Months from 1979)")
plt.title("u wind component - mean per longitude")
plt.show()
```
